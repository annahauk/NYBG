```python
# copy the images from the animal-specimens class to the folder called animal-specimens
df_animal = df_train[df_train["classLabel"] == "animal-specimens"]
```

```python
df_animal = df_animal["imageFile"]
#make it a dataframe
df_animal = pd.DataFrame(df_animal)
```

```python
df_animal
```

```python
import shutil
import os

source_directory = "/Users/annahauk/Desktop/NYBG/bttai-nybg-2024/BTTAIxNYBG-train/BTTAIxNYBG-train"
destination_directory = "/Users/annahauk/Desktop/NYBG/bttai-nybg-2024/animal-specimens"

for i in range(len(df_animal)):
    file_name = str(df_animal.iloc[i, 0])  # Assuming the file names are in the first column
    source_path = os.path.join(source_directory, file_name)
    destination_path = os.path.join(destination_directory, file_name)

    try:
        shutil.copyfile(source_path, destination_path)
        print(f"Successfully copied {file_name} to {destination_directory}")
    except FileNotFoundError:
        print(f"Error: {file_name} not found in {source_directory}")
    except Exception as e:
        print(f"Error: Unable to copy {file_name}. {e}")
```
